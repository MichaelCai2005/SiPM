```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math

data = np.loadtxt("Datasets\\50_10.2amplitude.csv", delimiter=",", skiprows=1)
x_values = data[:, 0]
y_values = data[:, 1]

# Define the superposition function
def spectrum_function(x, g, lambda_param, sigma0, S, B):
    N_max = 10  # Adjust based on your data
    spectrum = np.zeros_like(x)
    for N in range(N_max + 1):
        # Avoid division by zero for N=0
        sigma_N = sigma0 if N == 0 else sigma0 * np.sqrt(N)
        # Poisson amplitude
        A_N = S * (lambda_param**N * np.exp(-lambda_param)) / math.factorial(N)
        # Gaussian component
        spectrum += A_N * np.exp(- (x - N * g)**2 / (2 * sigma_N**2))
    return spectrum + B

# Initial parameter guesses
initial_guess = [250, 7, 25, 4000, 0]

# Curve fitting
popt, pcov = curve_fit(spectrum_function, x_values, y_values, p0=initial_guess)
g_fit, lambda_fit, sigma0_fit, S_fit, B_fit = popt
perr = np.sqrt(np.diag(pcov))  # Parameter uncertainties

# Print fitted parameters
print(f"Single-Photon Gain (g): {g_fit:.2f}  {perr[0]:.2f}")
print(f"Mean Photon Number (): {lambda_fit:.2f}  {perr[1]:.2f}")
print(f"Sigma at N=1 (): {sigma0_fit:.2f}  {perr[2]:.2f}")
print(f"Scale Factor (S): {S_fit:.2f}  {perr[3]:.2f}")
print(f"Baseline Offset (B): {B_fit:.2f}  {perr[4]:.2f}")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, 'b-', label="Experimental Data")
plt.plot(x_values, spectrum_function(x_values, *popt), 'r--', label="Superposition Fit")
plt.xlabel('ADC Channels')
plt.ylabel('Counts')
plt.title('LED Spectrum with Superposition of Gaussians Fit')
plt.legend()
plt.grid(True)
plt.show()
```
